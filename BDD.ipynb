## Étape 2 : séparation temporelle
Un véhicule envoie un paquet toutes les secondes. On peut donc considérer que si deux paquets consécutifs ont plus d'une seconde d'écart, c'est qu'ils appartiennent à deux échanges disctincts entre les deux mêmes véhicules.

```python
l=[]
for i in range(len(list_sender[1927][1984])-1):
    if (list_sender[1927][1984].iloc[i+1,7]-list_sender[1927][1984].iloc[i,7])>1:
        l.append(i)
for i in l:
    print("Paquet ",i," envoyé par 1927 et reçu par 1984, paquet ",list_sender[1927][1984].iloc[i+1,0], "de la BDD")
```

```python
database.iloc[29324:29334,1]
```

```python
list_sender[1927][1984].iloc[88:96,7]
```

On voit ici que certains paquets ont un écart de plus d'une seconde. Cependant cela peut correspondre a un paquet qui n'as pas été correctement envoyé ou reçu.

```python
l1=[]
for i in range(len(list_sender[1927][1984])-1):
    if (list_sender[1927][1984].iloc[i+1,7]<list_sender[1927][1984].iloc[i,7]):
        l1.append(i)
```

```python
l2=[]
for i in range(len(list_sender[1927][1984])-1):
    if (list_sender[1927][1984].iloc[i+1,2]<list_sender[1927][1984].iloc[i,2]):
        l2.append(i)
```
