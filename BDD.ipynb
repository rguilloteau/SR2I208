# Travail sur la base de données

```python
import os
import pandas as pd 
```

```python
L = [f for f in os.listdir('./Dataset.csv') if f[:6] == 'attack']
frames=[]
for el in L :
    frames.append(pd.read_csv('./Dataset.csv/'+el+''))
    
database = pd.concat(frames)
```

```python
database_sorted = dict()

for x in database.iterrows():
    if x['Identifiant émetteur'] in database_sorted:
        database_sorted[x['Identifiant émetteur']]+=x
    else:
        database_sorted[x['Identifiant émetteur']]=x
    
    
```

```python
list_sender = {x:database[database['Identifiant émetteur'] == x] for x in database['Identifiant émetteur'].unique()}
```
